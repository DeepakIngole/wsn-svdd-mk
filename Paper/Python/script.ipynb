```python
#!/usr/bin/env python
#-*- coding: utf-8 -*-
from __future__ import division
import matplotlib 
import scipy
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
#import python-mpi  todo

import subprocess
plt.style.use('classic')

## Initialize #{{{
matplotlib.rc('text', usetex=True)
matplotlib.rc('font',**{'family':'serif','serif':['Computer Modern Roman, Times']})
matplotlib.rc('text.latex',unicode=True)
colors = ("#BB3300", "#8800DD", "#2200FF", "#0099DD", "#00AA00", "#AA8800",
					   "#661100", "#440077", "#000088", "#003366", "#004400", "#554400")
#}}}

## Data
import scipy.io as sio
data=sio.loadmat('iddata.mat')
#}}}

plt.figure(figsize=(5,6))

## Ouput 1
ax3 = plt.subplot(2,1,1)
ax3.plot(data['freq'],data['magG'], color='red', label=u"$G ( z^{-1} )$", linewidth=2, linestyle=('-','--','-.',':')[0])
ax3.plot(data['freq'],data['magM'], color=colors[4], label=u"$M ( z^{-1} )$", linewidth=1.25, linestyle=('-','--','-.',':')[0])
ax3.plot(data['freq'],data['magWD'], color='black', label=u"$G_p ( z^{-1} )$", linewidth=1, linestyle=('-','--','-.',':')[0])
ax3.grid(True)
ax3.set_ylabel(u"Amplitude [dB]")
ax3.set_ylim((-60,40))
ax3.legend(loc=(.62,.5))
ax3.get_yaxis().set_label_coords(-0.12,0.5)
#}}}

## Ouput 2
ax4 = plt.subplot(2,1,2)
ax4.plot(data['freq'],data['magW'], color='red', label=u"$W ( z^{-1} )$", linewidth=2, linestyle=('-','--','-.',':')[0])
ax4.plot(data['freq'],data['magD'], color=colors[4], label=u"$D ( z^{-1} )$", linewidth=1.25, linestyle=('-','--','-.',':')[0])
ax4.grid(True)
ax4.set_ylabel(u"Amplitude [dB]")
ax4.set_xlabel(u"Frequency [Hz]")
ax4.set_ylim((-60,40))
ax4.legend(loc=(.62,.65))
ax4.get_yaxis().set_label_coords(-0.12,0.5)
#}}}

plt.savefig("system_identification.pdf", bbox_inches='tight')
plt.show()
```
