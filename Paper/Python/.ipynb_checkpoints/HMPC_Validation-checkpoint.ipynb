```python
#!/usr/bin/env python
#-*- coding: utf-8 -*-
from __future__ import division
import matplotlib 
import scipy
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
#import python-mpi  todo

import subprocess
plt.style.use('classic')

## Initialize #{{{
matplotlib.rc('text', usetex=True)
matplotlib.rc('font',**{'family':'serif','serif':['Computer Modern Roman, Times']})
matplotlib.rc('text.latex',unicode=True)
colors = ("#BB3300", "#8800DD", "#2200FF", "#0099DD", "#00AA00", "#AA8800",
					   "#661100", "#440077", "#000088", "#003366", "#004400", "#554400")
#}}}

## Data
import scipy.io as sio
data=sio.loadmat('valid_data1.mat')
Tsim=2000
#}}}

plt.figure(figsize=(10,6))

## Disturbance
ax1 = plt.subplot(2,2,1)
ax1.plot(data['t'],data['w1'], color='black', label=u"$NCR_{22}^d$", linewidth=1.5, linestyle=('-','--','-.',':')[0])
ax1.grid(True)
ax1.set_ylabel(u"Pulsed heat load [W]")
ax1.legend(loc=(.57,.75))
ax1.set_xlim((0,Tsim))
ax1.get_yaxis().set_label_coords(-0.12,0.5)
#}}}

## Control inputs
ax2 = plt.subplot(2,2,2)
ax2.plot(data['t'],data['u11'], color='red', label=u"$CV_{155}$", linewidth=1.5, linestyle=('-','--','-.',':')[0])
ax2.plot(data['t'],data['u12'], color=colors[4], label=u"$NCR_{22}^a$", linewidth=1.25, linestyle=('-','--','-.',':')[0])
ax2.plot(data['t'],data['u2'], color='black', label=u"$CV_{156}$", linewidth=1, linestyle=('-','--','-.',':')[0])
ax2.grid(True)
ax2.set_ylabel(u"Actuators [\%] or [W]")
ax2.set_ylim((0,70))
ax2.set_xlim((0,Tsim))
ax2.legend(loc=(.57,.1))
ax2.get_yaxis().set_label_coords(-0.12,0.5)
#}}}

## Ouput 1
ax3 = plt.subplot(2,2,3)
ax3.plot(data['t'],data['r1d'], color='red', label=u"$L_1^d$", linewidth=2, linestyle=('-','--','-.',':')[0])
ax3.plot(data['t'],data['r1'], color=colors[4], label=u"$L_1^a$", linewidth=1.25, linestyle=('-','--','-.',':')[0])
ax3.plot(data['t'],data['y1'], color='black', label=u"$L_1$", linewidth=1, linestyle=('-','--','-.',':')[0])
ax3.grid(True)
ax3.set_ylabel(u"Level bath [\%]")
ax3.set_xlabel(u"Time [s]")
ax3.set_xlim((0,Tsim))
ax3.set_ylim((51,66))
ax3.legend(loc=(.7,.1))
ax3.get_yaxis().set_label_coords(-0.12,0.5)
#}}}

## Ouput 2
ax4 = plt.subplot(2,2,4)
ax4.plot(data['t'],data['r2d'], color='red', label=u"$T_5^d$", linewidth=2, linestyle=('-','--','-.',':')[0])
ax4.plot(data['t'],data['r2'], color=colors[4], label=u"$T_5^a$", linewidth=1.25, linestyle=('-','--','-.',':')[0])
ax4.plot(data['t'],data['y2'], color='black', label=u"$T_5$", linewidth=1, linestyle=('-','--','-.',':')[0])
ax4.grid(True)
ax4.set_ylabel(u"Turbine temperature [K]")
ax4.set_xlabel(u"Time [s]")
ax4.set_xlim((0,Tsim))
ax4.set_ylim((6,13))
ax4.legend(loc=(.7,.1))
ax4.get_yaxis().set_label_coords(-0.12,0.5)
#}}}

plt.savefig("hmpc_valid1.pdf", bbox_inches='tight')
plt.show()
```
