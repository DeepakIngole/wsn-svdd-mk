```python
#!/usr/bin/env python
#-*- coding: utf-8 -*-
from __future__ import division
import matplotlib 
import scipy
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
#import python-mpi  todo

import subprocess
plt.style.use('classic')

## Initialize #{{{
matplotlib.rc('text', usetex=True)
matplotlib.rc('font',**{'family':'serif','serif':['Computer Modern Roman, Times']})
matplotlib.rc('text.latex',unicode=True)
colors = ("#BB3300", "#8800DD", "#2200FF", "#0099DD", "#00AA00", "#AA8800",
					   "#661100", "#440077", "#000088", "#003366", "#004400", "#554400")
#}}}

## Data
import scipy.io as sio
data0=sio.loadmat('boundary_data0.mat')
data1=sio.loadmat('boundary_data1.mat')
data2=sio.loadmat('boundary_data2.mat')
data3=sio.loadmat('boundary_data3.mat')
#}}}

plt.figure(figsize=(5,3))

## Ouput 1
ax = plt.subplot(1,1,1)
ax.plot(data1['positiveData'][:,0],data1['positiveData'][:,1],color=colors[4],linewidth=0,marker='o',markersize=3)
ax.plot(data0['boundaryData'][:,0],data0['boundaryData'][:,1],color='red',linewidth=0, marker='*',markersize=1,markeredgecolor='red')
ax.plot(data3['boundaryData'][:,0],data3['boundaryData'][:,1],color='red',linewidth=0, marker='*',markersize=1,markeredgecolor='green')
ax.plot(data1['boundaryData'][:,0],data1['boundaryData'][:,1],color='red',linewidth=0, marker='*',markersize=1,markeredgecolor='blue')
ax.plot(data2['boundaryData'][:,0],data2['boundaryData'][:,1],color='red',linewidth=0, marker='*',markersize=1,markeredgecolor='gray')
ax.grid(True)
ax.set_ylabel(u"Humidity [\%]")
ax.set_xlabel(u"Temperature [C]")
ax.set_xlim(14,34)
ax.set_ylim(15,60)
ax.get_yaxis().set_label_coords(-0.12,0.5)
#}}}

## Legend
import matplotlib.lines as mlines
import matplotlib.patches as mpatches

green_dot=mlines.Line2D([],[],color=colors[4],linewidth=0,marker='o',markersize=3,label="Samples")
blue_line=mlines.Line2D([],[],color='blue',label="$\delta=0.01$")
green_line=mlines.Line2D([],[],color='green',label="$\delta=0.02$")
red_line=mlines.Line2D([],[],color='red',label="$\delta=0.05$")
gray_line=mlines.Line2D([],[],color='gray',label="$\delta=0.1$")
ax.legend(loc=(.72,.54),fontsize=10,fancybox=True,handles=[green_dot,blue_line,green_line,red_line,gray_line])
#}}}

plt.savefig('data_description.pdf',bbox_inches='tight',rasterized=True)

plt.show()
```
